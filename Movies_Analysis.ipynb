```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= {'Title': ['The Shawshank Redemption','The Godfather','The Dark Knight','Pulp Fiction',
              'Forrest Gump','Inception','The Matrix','Interstellar','Goodfellas',
              'Fight Club','The Silence of the Lambs','Schindler List','The Lord of the Rings',
              'Star Wars','Jurassic Park','Titanic','Avatar','The Lion King',
              'Home Alone','Die Hard'],
    'Year': [1994,1972,2008,1994,1994,2010,1999,2014,1990,1999,
             1991,1993,2003,1977,1993,1997,2009,1994,1990,1988],
    'Genre': ['Drama','Crime','Action','Crime','Drama','Sci-Fi','Sci-Fi','Sci-Fi','Crime','Drama',
              'Thriller','Drama','Fantasy','Sci-Fi','Adventure','Romance','Sci-Fi','Animation',
              'Comedy','Action'],
    'Rating': [9.3,9.2,9.0,8.9,8.8,8.8,8.7,8.6,8.7,8.8,
               8.6,8.9,8.9,8.6,8.1,7.8,7.9,8.5,7.6,8.2],
    'Budget_M': [25,6,185,8,55,160,63,165,25,63,
                 19,22,94,11,63,200,237,45,18,28],
    'Revenue_M': [73,246,1005,214,678,836,467,773,46,101,
                  273,322,1119,775,1030,2187,2923,968,476,140],
    'Duration_Min': [142,175,152,154,142,148,136,169,146,139,
                     118,195,201,121,127,194,162,88,103,132]
}
df= pd.DataFrame(data)
```
